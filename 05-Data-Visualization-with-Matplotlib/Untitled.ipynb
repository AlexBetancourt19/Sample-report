```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional method
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#if you are not in JN, use .plot()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

```python
#OBJECT ORIENTED:
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger Title')

axes2.plot(y,x)
axes2.set_title('Smaller Title')
```

```python
#part 2:
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

#for current_ax in axes:
#    current_ax.plot(x, y)

axes[0].plot(x,y)
axes[0].set_title('First title')

axes[1].plot(y,x)
axes[1].set_title('Second title')
```

```python
##figure size and dpi
#dpi stands for "Dots per inches"
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=100)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X squared')
ax.plot(x,x**3, label='X cubed')

ax.legend(loc=0)
#0 is the best location 
#loc is legend location, see reference for more details

#.loc(x, y) for customized location
```

```python
#part 3- Plot appearance
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00')
ax.plot(x,x**3,color='Blue',linewidth=5)
ax.plot(x,x**4,color='Red',alpha=0.5)
ax.plot(x,x**5,color='Green',ls='--')
```

```python
fig, axes = plt.subplots(6,1,figsize=(8,10))
axes[0].plot(x,y,color='#FF8C00')
axes[1].plot(x,y,color='Red')
axes[2].plot(x,y,color='Orange',linewidth=5)
axes[3].plot(x,y,color='Blue',ls='--')
#see notes for more reference for linestyle
axes[4].plot(x,y,color='Green',ls='-',marker=6,markersize=10)
#for marker references see notes
```
